```python
from typing_extensions import Required
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import requests
```

```python
url_dataset = "https://raw.githubusercontent.com/Mart1n2015/MartinCerioli/2ecf72ab3b51ec7a33bdcdcad861580635352f5e/Datos%20hist%C3%B3ricos%20de%20Tenaris%20ADR%20(TS).csv"
```

```python
df = pd.read_csv(url_dataset,sep = ";")
```

```python
df
```

```python
df['PETROLEO'] = df['PETROLEO'].str.replace(',', '.').astype(float)
df['TS'] = df['TS'].str.replace(',', '.').astype(float)
df['DÓLAR'] = df['DÓLAR'].str.replace(',', '.').astype(float)
df['S&P'] = df['S&P'].str.replace(',', '.').astype(float)
```

```python
x1 = df ['PETROLEO']
x2 = df ['DÓLAR']
x3 = df ['S&P']
y = df ['TS']
```

```python
modelo = LinearRegression()
X = pd.concat([x1, x2, x3], axis=1)
modelo.fit(X, y)
```

```python
print(f"Coeficiente (pendiente): {modelo.coef_[0]:.4f}")
print(f"Intercepto: {modelo.intercept_:.4f}")
print(f"R² (bondad del ajuste): {modelo.score(X, y):.4f}")
```
